```python
import pandas as pd
from sqlalchemy import create_engine
from datetime import datetime
```

```python
spotify_file = "Resources/top2018.csv"
spotify_df = pd.read_csv(spotify_file)
spotify_df.head()
```

```python
billboard_file = "Resources/Hot Stuff.csv"
billboard_df = pd.read_csv(billboard_file)
billboard_df.head()
```

```python
spotify_df = spotify_df[["name", "artists", "duration_ms"]]
spotify_df['name'] = spotify_df['name'].str.split('(', 1).str[0].str.strip()
spotify_df["position"] = (spotify_df.index + 1)
spotify_df = spotify_df.rename(columns={"position":"spotify_position", "name":"song_name"})
spotify_df.set_index("spotify_position", inplace=True)
spotify_df.head()
```

```python
billboard_df['WeekID'] =  pd.to_datetime(billboard_df['WeekID'], format= '%m/%d/%Y')

billboard_2018_df = billboard_df[(billboard_df['WeekID'].dt.year > 2017)]
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
billboard_2018_df['WeekID'] = billboard_2018_df['WeekID'].dt.strftime('%m/%d/%Y')
billboard_2018_df.head(50)
billboard_2018_df.sort_values(by='WeekID', ascending=False)
```

```python
billboard_2018_df = billboard_2018_df[["Song", "Performer", "WeekID", "Week Position", "Peak Position", "Weeks on Chart"]]
# billboard_2018_df.head()
by_song = billboard_2018_df.groupby(["Song"])
by_song.count()

num_weeks = by_song["Weeks on Chart"].max()
avg_position = by_song["Week Position"].mean()
peak_position = by_song["Peak Position"].min()

billboard_summary = pd.DataFrame({"Number of Weeks":num_weeks,"Average Chart Position": avg_position, 
                                  "Peak Position": peak_position})

billboard_summary.sort_values(by='Number of Weeks', ascending=False)
# billboard_summary.head()


billboard_merge = pd.merge(billboard_2018_df, billboard_summary, on="Song")
# billboard_merge.head()
```

```python
billboard_final = billboard_merge[["Song","Performer","Number of Weeks", "Average Chart Position", "Peak Position_y"]]
billboard_final.drop_duplicates("Song", inplace=True)
billboard_final = billboard_final.rename(columns={"Peak Position_y":"peak_position", "Song":"song_name", "Performer":"artists", "Number of Weeks":"number_of_weeks", "Average Chart Position":"average_chart_position"})
billboard_final.sort_values(by=['peak_position','number_of_weeks'], ascending=[True, False], inplace=True)
# billboard_final.set_index("peak_position", inplace=True)
billboard_final.reset_index(drop=True, inplace=True)
billboard_final.head(100)
```
